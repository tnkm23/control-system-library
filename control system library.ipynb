```python
import numpy as np
import matplotlib.pyplot as plt

K = 1.0  # ゲイン
T = 2.0  # 時定数

# 応答の時間範囲を指定
t = np.linspace(0, 10, 1000)

# ステップ入力信号の定義
u = np.ones_like(t)

# 1次遅れ応答の計算
y = K * (1 - np.exp(-t/T))

# グラフのプロット
plt.plot(t, u, 'b-', label='Step Input')
plt.plot(t, y, 'r-', label='Step Response')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.title('Step Response of First-Order System')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import control
import matplotlib.pyplot as plt

# システムの伝達関数を定義
L = 2.0  # むだ時間
T = 5.0  # 時定数
K = 1.0  # ゲイン
sys = control.tf([K], [T, 1])

# 遅延項を近似
num, den = sys.num[0][0], sys.den[0][0]
num_pade, den_pade = control.pade(L, 5)  # 5次近似
num_delayed = np.convolve(num, num_pade)
den_delayed = np.convolve(den, den_pade)
sys_with_delay = control.tf(num_delayed, den_delayed)

# ステップ応答を計算
t = np.linspace(0, 20, 1000)
t, y = control.step_response(sys_with_delay, T=t)

# グラフのプロット
plt.plot(t, y)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.title('Step Response of First-Order System with Time Delay')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control

K = 1  # ゲイン
T = 2  # 時定数

# 応答の時間範囲を指定
t1 = np.linspace(0, 14, 1000)

# ステップ入力信号の定義
u = np.ones_like(t1)

# 1次遅れ応答の計算
y = K * (1 - np.exp(-t1/T))


#cotrol system library で実装
num = np.array([K])
den = np.array([T, 1])

H = control.tf(num, den)
print('H(s) =', H)

t2, y2 = control.step_response(H)

# グラフのプロット
plt.plot(t1, u, 'b-', label='Step Input')
plt.plot(t1, y, 'r-', label='Step Response')
plt.plot(t2, y2, 'g-', label='cotrol system library')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.title('Step Response of First-Order System')
plt.legend()
plt.grid()
plt.show()
```
